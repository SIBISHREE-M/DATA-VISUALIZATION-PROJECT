```python
import pandas as pd

# Load the dataset
data = pd.read_csv('/kaggle/input/london-weather-data/london_weather.csv')

# Check the first few rows of the dataset
print(data.head())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(data['date'], data['mean_temp'], label='Mean Temperature')
plt.xlabel('Date')
plt.ylabel('Mean Temperature (Â°C)')
plt.title('Mean Temperature Over Time')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(data['mean_temp'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Mean Temperature (Â°C)')
plt.ylabel('Frequency')
plt.title('Mean Temperature Distribution')
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('/kaggle/input/london-weather-data/london_weather.csv')

# Select the columns for correlation analysis
correlation_data = data[['mean_temp', 'precipitation', 'sunshine', 'global_radiation']]

# Calculate the correlation matrix
correlation_matrix = correlation_data.corr()

# Create a heatmap for the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(data['mean_temp'], data['precipitation'])
plt.xlabel('Mean Temperature (Â°C)')
plt.ylabel('Precipitation (mm)')
plt.title('Mean Temperature vs. Precipitation')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('/kaggle/input/london-weather-data/london_weather.csv')

# Create the heatmap using date and mean temperature
heatmap_data = data.pivot_table(index='date', values='mean_temp')
plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data, cmap='coolwarm', cbar=True)
plt.title('Mean Temperature Heat Map')
plt.show()
```
