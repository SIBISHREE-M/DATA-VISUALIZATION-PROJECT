```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.pyplot as plt  # Add this import statement

# Load the dataset
data = pd.read_csv('/kaggle/input/flipkart-product-dataset/train.csv')  # Update with your dataset path

# Identify top products based on the number of reviews and ratings
top_products = data.nlargest(10, 'noreviews1')

# Create a bar chart for the top products
plt.figure(figsize=(10, 6))
plt.barh(top_products['title'], top_products['noreviews1'])
plt.xlabel('Number of Reviews')
plt.ylabel('Product Title')
plt.title('Top 10 Products with the Most Reviews')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('/kaggle/input/flipkart-product-dataset/train.csv')  # Update with your dataset path

# Analyze user satisfaction based on product ratings
plt.figure(figsize=(10, 6))
sns.histplot(data['Rating'], bins=5, kde=True)

plt.xlabel('Ratings')
plt.ylabel('Count')
plt.title('Distribution of User Ratings')
plt.show()
```

```python
popular_products = data['id'].value_counts().head(10)
plt.figure(figsize=(12,6))
popular_products.plot(kind='bar', color='skyblue')
plt.xlabel('Product ID')
plt.ylabel('Frequency')
plt.title('Top 10 Most Popular Products')
plt.show()
```

```python

popular_categories = data['title'].value_counts().head(10)
plt.figure(figsize=(10,10))
plt.pie(popular_categories, labels=popular_categories.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 10 Most Popular Categories')
plt.show()
```

```python
import plotly.graph_objects as go
fig = go.Figure(go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color="black", width=0.5),
        label=["Product A", "Product B", "Product C", "Product D"],
    ),
    link=dict(
        source=[0, 1, 2, 0, 3],  # indices correspond to labels
        target=[2, 3, 3, 1, 2],
        value=[8, 4, 2, 8, 2],
    )
))

fig.update_layout(title_text="User Click Patterns - Sankey Diagram", font_size=10)
fig.show()
```
