```python
import pandas as pd

# Load your dataset into a DataFrame
df = pd.read_csv('/kaggle/input/us-census-demographic-data/acs2015_census_tract_data.csv')

# Now you can use the 'df' DataFrame for your visualization code
import matplotlib.pyplot as plt

# Group data by state and calculate the mean Native population percentage
state_native_mean = df.groupby('State')['Native'].mean().sort_values()

plt.bar(state_native_mean.index, state_native_mean, color='green')
plt.xlabel('State')
plt.ylabel('Mean % Native Population')
plt.title('Mean Native Population Percentage by State')
plt.xticks(rotation=90)  # Rotate x-axis labels for readability
plt.show()
```
